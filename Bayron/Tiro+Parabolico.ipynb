# Tiro Parabólico
v = V0cos(θ)i + V0sen(θ)j
a = -gj
x= (V0xt + X0)i + (-1/2gt^2 + V0yt+Y0)j

```julia
t = collect(0:0.125:20)
```

```julia
g=9.81 # m/s^2
```

```julia
v_0 = 100 #m/s
```

```julia
θ = π/4
```

```julia
v = v_0*cos(θ)+v_0*sin(0)
```

```julia
a = -g
```

```julia
X=zeros(t);
Y=zeros(t);
```

```julia
t[132]
```

```julia
for i ∈ eachindex(t)
    X[i] = v_0*cos(θ)*t[i]
    Y[i] = (-(0.5*g*t[i]^2)+v_0*sin(θ)*t[i])
end
```

```julia
Pkg.add("Plots")
```

```julia
using Plots
```

```julia
pyplot()
```

```julia
plot(X,Y,w=2)
title!("Tiro parabolico")
xaxis!("posicion en x (m)")
yaxis!("posicion en y (m)")
```
